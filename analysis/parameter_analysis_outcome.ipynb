### Graph 2: statistical comparison of recurrence indices between outcome groups
Get p-values by parameter combination for variable you want to predict

```python
df = get_p_vals(taus, ms, coefs, feat_name, y_proc, step)
p_vals = df['p_val']
p_vals = np.hstack(p_vals)
scores = -np.log10(p_vals)
sig_level_5 = -np.log10(0.05) / scores.max()
sig_level_1 = -np.log10(0.01) / scores.max()
scores /= scores.max()
df['scores'] = scores
```

```python
f = plt.figure(figsize=(30, 20))
sns.set_style('white')
sns.set_context('poster')
x = df[['tau', 'm', 'coef']].apply(tuple,axis=1)
x.name = 'tau, m, coef'
e = sns.barplot(x=x, y='scores', data=df, color=sns.color_palette()[0])
e.set_xlabel('tau, m, coef', fontsize=60)
e.set_ylabel(r'$\mathregular{\frac{-log_{10}(p)}{max(-log_{10}(p))}}$', fontsize=70)
e.set(ylim=(0, 1))
e.set_title(feat_name + ' statistical comparison', fontsize=50)
e.tick_params(axis='both', which='major', labelsize=50)

if sig_level_5 < 1:
    plt.axhline(y=sig_level_5, color='k', linestyle=':')
    plt.text(6, sig_level_5 + 0.01, 'significance level, p < 0.05', fontsize=50)
if sig_level_1 < 1:
    plt.axhline(y=sig_level_1, color='k', linestyle=':')
    plt.text(6, sig_level_1 + 0.01, 'significance level, p < 0.01', fontsize=50)
rotate_xtick = plt.xticks(rotation=45)   # assign to variable just to suppress output

sns.despine()
```
